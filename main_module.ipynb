```python
#### from car_rental import CarRental, Customer
from datetime import datetime

def main():
    rental = CarRental()

    while True:
        print("\nüöó Welcome to the Car Rental Service!")
        print("1Ô∏è‚É£  Display available cars")
        print("2Ô∏è‚É£  Rent a car")
        print("3Ô∏è‚É£  Return a car")
        print("4Ô∏è‚É£  Exit")

        choice = input("\nEnter your choice: ").strip()

        if choice == '1':
            rental.display_available_cars()

        elif choice == '2':
            name = input("Enter your name: ").strip()
            email = input("Enter your email: ").strip()
            customer = Customer(name, email)

            vehicle_type = input("Enter vehicle type (Sedan/SUV/Hatchback): ").strip()
            try:
                num_cars = int(input("Enter number of cars: "))
            except ValueError:
                print("‚ùå Please enter a valid number.")
                continue

            rental_period = input("Enter rental period (hourly/daily/weekly): ").strip().lower()
            rental_time = datetime.now()

            if rental_period == 'hourly':
                rental.rent_hourly(vehicle_type, num_cars, rental_time, customer)
            elif rental_period == 'daily':
                rental.rent_daily(vehicle_type, num_cars, rental_time, customer)
            elif rental_period == 'weekly':
                rental.rent_weekly(vehicle_type, num_cars, rental_time, customer)
            else:
                print("‚ö†Ô∏è Invalid rental period! Please choose hourly, daily, or weekly.")

        elif choice == '3':
            print("\nProcessing returns...")
            rental.return_cars(datetime.now())

        elif choice == '4':
            print("\nüôè Thank you for using our Car Rental Service. Goodbye!")
            break

        else:
            print("‚ö†Ô∏è Invalid choice. Please choose again (1‚Äì4).")

if __name__ == "__main__":
    main()
```
